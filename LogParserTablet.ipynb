```python
# look for how many times the light intensity was changed(tablet)
LightIntensity = 0
lastTime = 0
lastDay= 0

if fileTablet.find('tablet') !=-1:
    for num4, strline in enumerate(fileListTablet):
        if len(strline) == 0:
            # do nothing
            x = 1
        # ['intensityChanged', 'Fri', 'Mar', '31', '12:03:47', '2017']
        if 'intensityChanged' in strline:
            t = strline[4]
            (h, m, s) = t.split(':')
            result = int(h) * 3600 + int(m) * 60 + int(s)
            #print('time in seconds: ', result)
            if lastTime == 0:
                lastDay = strline[1]
                lastTime = result
                print('light intensity changed for first time in file: ', strline[4], ' day: ', lastDay)
            elif (result - lastTime) < 60 and (result - lastTime) > -1:
                lastDay = strline[1]
                lastTime = result
            else: # count at end of movement
                LightIntensity = LightIntensity + 1
                print('light intensity ++: ', lastTime, ' day: ', lastDay)
                lastDay = strline[3]
                lastTime = result

print('Light intensity: ', LightIntensity+1) # will not have counted last one
```

```python
# look for how many times the lights were moved (tablet)
LightMoved = 0
lastResult = 0
lastDay = 0
lastTime = 0

listOfMovementsTabletX = []
listOfMovementsTabletY = []
listOfMovementsTabletT = []
tempListX = []
tempListY = []
tempListT = []

if fileTablet.find('tablet') !=-1:
    for num5, strline in enumerate(fileListTablet):
        if len(strline) == 0:
            # do nothing
            x = 1
        # ['positionChanged', 'Fri', 'Mar', '31', '12:03:40', '2017']
        elif 'positionChanged' in strline:
            #print('position changed: ', strline[1], ' time: ', strline[4], ' dif: ', (result-lastTime) )
            t = strline[4]
            (h, m, s) = t.split(':')
            result = int(h) * 3600 + int(m) * 60 + int(s)
            #print('time in seconds: ', result)
            if lastTime == 0:
                lastDay = strline[1]
                lastResult = result
                lastTime = strline[4]
                print('light moved for first time in file: ', lastTime)
            elif (result - lastResult) < 20 and (result - lastResult) > -1:
                lastDay = strline[1]
                lastResult = result
                lastTime = strline[4]
                #print('still in same movement')
            else: 
                listOfMovementsTabletX.append(tempListX)
                listOfMovementsTabletY.append(tempListY)
                listOfMovementsTabletT.append(tempListT)
                # clear lists
                tempListX = []
                tempListY = []
                tempListT = []
                # count at end of movement
                LightMoved = LightMoved + 1
                print('light moved ++, day: ', lastDay, ' time: ',  lastTime)
                lastDay = strline[1]
                lastResult = result
                lastTime = strline[4]
            tempListT.append(result)
        if 'pos:' in strline:
            tempListX.append(strline[1])
            tempListY.append(strline[2])

listOfMovementsTabletX.append(tempListX)
listOfMovementsTabletY.append(tempListY)
listOfMovementsTabletT.append(tempListT)

print('light moved ++, day: ', lastDay, ' time: ',  lastTime)
print('Lights moved: ', LightMoved+1) # will not have counted last one
print('Length of movement list X: ', len(listOfMovementsTabletX))
print('Length of movement list T: ', len(listOfMovementsTabletT))
```
